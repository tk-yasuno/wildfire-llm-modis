import requests
import json
import pandas as pd
from datetime import datetime

url = "https://modis.ornl.gov/rst/api/v1/"
header = {'Accept': 'application/json'} # Use following for a csv response: header = {'Accept': 'text/csv'}

csv = "example_sites.csv"
First, read your file into a pandas data frame:

coordinates = pd.read_csv(csv)
print(coordinates)

---

Before we can submit orders for the global tool, we must find the MODIS dates nearest to the start and end of our desired time series. We will use the dates function to get a list of available MODIS dates for each of our input coordinates:

/api/v1/{product}/dates
The dates function returns a list of available dates for the specified coordinate and MODIS product.

Parameter	Description
product	MODIS product code as listed by products function
latitude	latitude
longitude	longitude
Iterate through the coordinates in the data frame and find the MODIS date nearest to the calendar dates in the start_date and end_date columns. Add them to a new column of the data frame:

#### Convert start_date and end_date columns to datetimes
coordinates['start_date'] =  pd.to_datetime(coordinates['start_date'])
coordinates['end_date'] =  pd.to_datetime(coordinates['end_date'])

#### Make new columns for MODIS start and end dates 
coordinates['start_MODIS_date'] = '' 
coordinates['end_MODIS_date'] = ''

for index, row in coordinates.iterrows():
    # Submit request
    response = requests.get('https://modis.ornl.gov/rst/api/v1/' + row['product'] + '/dates?latitude=' + str(row['latitude']) + '&longitude='+ str(row['longitude']), headers=header)
    
    # Get dates object as list of python dictionaries
    dates = json.loads(response.text)['dates'] 
    
    # Convert to list of tuples; change calendar_date key values to datetimes
    dates = [(datetime.strptime(date['calendar_date'], "%Y-%m-%d"), date['modis_date']) for date in dates]
    
    # Get MODIS dates nearest to start_date and end_date and add to new pandas columns
    coordinates.loc[index, 'start_MODIS_date'] = min(date[1] for date in dates if date[0] > row['start_date'])
    coordinates.loc[index, 'end_MODIS_date'] = max(date[1] for date in dates if date[0] < row['end_date'])

print(coordinates)

--- 

Now, we are ready to submit our subset orders. We will use the subsetOrder function to pass our subset parameters to the ORNL DAAC's MODIS Global Subset Tool service:

/api/v1/{product}/subsetOrder
The subsetOrder function returns a unique order identifier (uid) that you can use to retreive your order URL. You will also receive an email at the supplied email address once the order has completed processing. Processing times vary based on the size of the order. Most are completed within 30 minutes.

Parameter	Description
product	MODIS product code as listed by products function
latitude	latitude
longitude	longitude
email	email address for order delivery
uid	unique order identifier
start_date	MODIS start date as listed by dates ("AYYYYDOY")
end_date	MODIS end date as listed by dates ("AYYYYDOY")
kmAboveBelow	number of kilometers to subset above and below center pixel
kmLeftRight	number of kilometers to subset left and right of center pixel
Iterate again through the rows of the dataframe and submit orders using the subsetOrder function:

#### Make list to collect order UIDs
order_uids = []

for index, row in coordinates.iterrows():
    # Build request URL
    requestURL = url + row['product'] + "/subsetOrder?latitude=" + str(row['latitude']) + "&longitude=" + str(row['longitude']) + "&email=" + row['email'] + "&uid=" + row['site_id'] + "&startDate=" + row['start_MODIS_date'] + "&endDate=" + row['end_MODIS_date'] + "&kmAboveBelow=" + str(row['kmAboveBelow']) + "&kmLeftRight=" + str(row['kmLeftRight'])

    # Submit request
    response = requests.get(requestURL, headers=header)
    
    # Append UID to list
    order_uids.append(json.loads(response.text)['order_id'])
    
print(order_uids)

## 4.MODIS Global Subset Tool API ãƒ‡ãƒ¢å®Ÿè£…
### NASA ORNL DAAC ã®MODIS Web Service ã‚’ä½¿ç”¨ã—ãŸå®Ÿéš›ã®ãƒ‡ãƒ¼ã‚¿å–å¾—

````python
# MODIS Global Subset Tool API ãƒ‡ãƒ¢å®Ÿè£…
# NASA ORNL DAAC ã®MODIS Web Service ã‚’ä½¿ç”¨ã—ãŸå®Ÿéš›ã®ãƒ‡ãƒ¼ã‚¿å–å¾—

import requests
import json
import pandas as pd
from datetime import datetime, timedelta
import time

print("ğŸ“š å¿…è¦ãªãƒ©ã‚¤ãƒ–ãƒ©ãƒªã‚’ã‚¤ãƒ³ãƒãƒ¼ãƒˆä¸­...")
print(f"âœ… requests: {requests.__version__}")
print(f"âœ… pandas: {pd.__version__}")
print(f"âœ… åŸºæœ¬ãƒ©ã‚¤ãƒ–ãƒ©ãƒªã®æº–å‚™å®Œäº†")

class MODISAPIClient:
    """MODIS Global Subset Tool API ã‚¯ãƒ©ã‚¤ã‚¢ãƒ³ãƒˆ"""
    
    def __init__(self):
        self.base_url = "https://modis.ornl.gov/rst/api/v1/"
        self.headers = {'Accept': 'application/json'}
        
    def get_available_products(self):
        """åˆ©ç”¨å¯èƒ½ãªMODISè£½å“ã®ä¸€è¦§ã‚’å–å¾—"""
        try:
            print("ğŸŒ MODIS API ã«æ¥ç¶šä¸­...")
            response = requests.get(f"{self.base_url}products", headers=self.headers, timeout=30)
            
            print(f"ğŸ“¡ HTTP Status: {response.status_code}")
            print(f"ğŸ“„ Response Headers: {dict(response.headers)}")
            
            if response.status_code == 200:
                try:
                    products = response.json()
                    print(f"âœ… JSONè§£ææˆåŠŸ")
                    print(f"âœ… åˆ©ç”¨å¯èƒ½ãªMODISè£½å“ãƒ‡ãƒ¼ã‚¿å–å¾—å®Œäº†")
                    return products
                except json.JSONDecodeError as e:
                    print(f"âŒ JSONè§£æã‚¨ãƒ©ãƒ¼: {e}")
                    print(f"ğŸ“„ Raw Response: {response.text[:500]}")
                    return None
            else:
                print(f"âŒ è£½å“ä¸€è¦§å–å¾—ã‚¨ãƒ©ãƒ¼: {response.status_code}")
                print(f"ğŸ“„ Error Response: {response.text[:300]}")
                return None
        except requests.exceptions.Timeout:
            print(f"âŒ ã‚¿ã‚¤ãƒ ã‚¢ã‚¦ãƒˆã‚¨ãƒ©ãƒ¼: APIã‚µãƒ¼ãƒãƒ¼ã®å¿œç­”ãŒé…ã™ãã¾ã™")
            return None
        except requests.exceptions.ConnectionError:
            print(f"âŒ æ¥ç¶šã‚¨ãƒ©ãƒ¼: APIã‚µãƒ¼ãƒãƒ¼ã«æ¥ç¶šã§ãã¾ã›ã‚“")
            return None
        except Exception as e:
            print(f"âŒ äºˆæœŸã—ãªã„ã‚¨ãƒ©ãƒ¼: {e}")
            return None
    
    def get_available_dates(self, product, latitude, longitude):
        """æŒ‡å®šåº§æ¨™ãƒ»è£½å“ã®åˆ©ç”¨å¯èƒ½æ—¥ä»˜ã‚’å–å¾—"""
        try:
            url = f"{self.base_url}{product}/dates"
            params = {'latitude': latitude, 'longitude': longitude}
            
            print(f"ğŸ“… {product} ã®åˆ©ç”¨å¯èƒ½æ—¥ä»˜ã‚’å–å¾—ä¸­... ({latitude}, {longitude})")
            response = requests.get(url, params=params, headers=self.headers, timeout=30)
            
            if response.status_code == 200:
                dates_data = response.json()
                dates = dates_data.get('dates', [])
                print(f"âœ… åˆ©ç”¨å¯èƒ½æ—¥ä»˜: {len(dates)} ä»¶")
                return dates
            else:
                print(f"âŒ æ—¥ä»˜å–å¾—ã‚¨ãƒ©ãƒ¼: {response.status_code}")
                print(f"Response: {response.text[:200]}")
                return None
        except Exception as e:
            print(f"âŒ æ—¥ä»˜å–å¾—ã‚¨ãƒ©ãƒ¼: {e}")
            return None
    
    def get_subset_data(self, product, latitude, longitude, start_date=None, end_date=None, km_above_below=1, km_left_right=1):
        """è¡›æ˜Ÿãƒ‡ãƒ¼ã‚¿ã®ã‚µãƒ–ã‚»ãƒƒãƒˆã‚’å–å¾—"""
        try:
            url = f"{self.base_url}{product}/subset"
            
            # ãƒ‡ãƒ•ã‚©ãƒ«ãƒˆã®æ—¥ä»˜è¨­å®šï¼ˆæœ€è¿‘ã®1é€±é–“ï¼‰
            if not start_date or not end_date:
                end_dt = datetime.now()
                start_dt = end_dt - timedelta(days=7)
                start_date = start_dt.strftime("%Y-%m-%d")
                end_date = end_dt.strftime("%Y-%m-%d")
            
            params = {
                'latitude': latitude,
                'longitude': longitude,
                'startDate': start_date,
                'endDate': end_date,
                'kmAboveBelow': km_above_below,
                'kmLeftRight': km_left_right
            }
            
            print(f"ğŸ›°ï¸ {product} ã‚µãƒ–ã‚»ãƒƒãƒˆãƒ‡ãƒ¼ã‚¿å–å¾—ä¸­...")
            print(f"ğŸ“ åº§æ¨™: ({latitude}, {longitude})")
            print(f"ğŸ“… æœŸé–“: {start_date} ~ {end_date}")
            
            response = requests.get(url, params=params, headers=self.headers, timeout=60)
            
            if response.status_code == 200:
                subset_data = response.json()
                print(f"âœ… ã‚µãƒ–ã‚»ãƒƒãƒˆãƒ‡ãƒ¼ã‚¿å–å¾—æˆåŠŸ")
                return subset_data
            else:
                print(f"âŒ ã‚µãƒ–ã‚»ãƒƒãƒˆå–å¾—ã‚¨ãƒ©ãƒ¼: {response.status_code}")
                print(f"Response: {response.text[:500]}")
                return None
                
        except Exception as e:
            print(f"âŒ ã‚µãƒ–ã‚»ãƒƒãƒˆå–å¾—ã‚¨ãƒ©ãƒ¼: {e}")
            return None

# MODIS APIã‚¯ãƒ©ã‚¤ã‚¢ãƒ³ãƒˆã‚’åˆæœŸåŒ–
modis_client = MODISAPIClient()

print("ğŸ” MODIS Global Subset Tool API ãƒ‡ãƒ¢ã‚’é–‹å§‹...")
print("="*60)

# 1. åˆ©ç”¨å¯èƒ½ãªè£½å“ä¸€è¦§ã‚’å–å¾—
print("\nğŸ“‹ Step 1: åˆ©ç”¨å¯èƒ½ãªMODISè£½å“ã‚’ç¢ºèª")
products = modis_client.get_available_products()

if products:
    # ãƒ‡ãƒãƒƒã‚°: ãƒ¬ã‚¹ãƒãƒ³ã‚¹å½¢å¼ã‚’ç¢ºèª
    print(f"ğŸ” DEBUG: products ã®å‹: {type(products)}")
    print(f"ğŸ” DEBUG: products ã®å†…å®¹ï¼ˆæœ€åˆã®200æ–‡å­—ï¼‰: {str(products)[:200]}")
    
    # ãƒ¬ã‚¹ãƒãƒ³ã‚¹å½¢å¼ã«å¿œã˜ã¦å‡¦ç†ã‚’åˆ†å²
    fire_related_products = []
    
    if isinstance(products, dict):
        # è¾æ›¸å½¢å¼ã®å ´åˆ
        if 'products' in products:
            products_list = products['products']
        else:
            products_list = [products]  # å˜ä¸€è£½å“ã®å ´åˆ
    elif isinstance(products, list):
        # ãƒªã‚¹ãƒˆå½¢å¼ã®å ´åˆ
        products_list = products
    else:
        print(f"âŒ äºˆæœŸã—ãªã„ãƒ¬ã‚¹ãƒãƒ³ã‚¹å½¢å¼: {type(products)}")
        products_list = []
    
    print(f"ğŸ“Š å‡¦ç†å¯¾è±¡è£½å“æ•°: {len(products_list)}")
    
    # ç«ç½æ¤œçŸ¥ã«é–¢é€£ã™ã‚‹è£½å“ã‚’æŠ½å‡º
    for product in products_list:
        try:
            if isinstance(product, dict):
                product_name = product.get('product', '')
                description = product.get('description', '').lower()
                
                # ç«ç½ã€æ¤ç”Ÿã€åœ°è¡¨æ¸©åº¦é–¢é€£ã®è£½å“ã‚’ç‰¹å®š
                if any(keyword in description for keyword in ['fire', 'temperature', 'vegetation', 'thermal', 'lst']):
                    fire_related_products.append(product)
                    
            elif isinstance(product, str):
                # è£½å“åã®ã¿ã®å ´åˆ
                product_name = product.lower()
                if any(keyword in product_name for keyword in ['fire', 'temp', 'vegetation', 'thermal', 'lst', 'mod14', 'mod13', 'mod11']):
                    fire_related_products.append({'product': product, 'description': 'Product name only'})
                    
        except Exception as e:
            print(f"âš ï¸ è£½å“å‡¦ç†ã‚¨ãƒ©ãƒ¼: {e}, è£½å“: {product}")
    
    print(f"\nğŸ”¥ ç«ç½æ¤œçŸ¥é–¢é€£è£½å“ ({len(fire_related_products)} ç¨®é¡):")
    for i, product in enumerate(fire_related_products[:5], 1):  # æœ€åˆã®5ã¤ã‚’è¡¨ç¤º
        if isinstance(product, dict):
            print(f"{i}. {product.get('product', 'N/A')}: {product.get('description', 'N/A')}")
            print(f"   è§£åƒåº¦: {product.get('resolution_meters', 'N/A')}m, é »åº¦: {product.get('frequency', 'N/A')}")
        else:
            print(f"{i}. {product}")
    
    if len(fire_related_products) == 0:
        print("âš ï¸ ç«ç½é–¢é€£è£½å“ãŒè¦‹ã¤ã‹ã‚Šã¾ã›ã‚“ã§ã—ãŸã€‚åˆ©ç”¨å¯èƒ½ãªå…¨è£½å“ã‚’è¡¨ç¤º:")
        for i, product in enumerate(products_list[:10], 1):  # æœ€åˆã®10å€‹ã‚’è¡¨ç¤º
            if isinstance(product, dict):
                print(f"{i}. {product.get('product', 'N/A')}: {product.get('description', 'N/A')[:100]}...")
            else:
                print(f"{i}. {product}")
else:
    print("âŒ è£½å“ãƒ‡ãƒ¼ã‚¿ã®å–å¾—ã«å¤±æ•—ã—ã¾ã—ãŸ")

print("\n" + "="*60)
print("âœ… MODIS API è£½å“ç¢ºèªå®Œäº†")
````

````python
# çµ±åˆç«ç½æ¤œçŸ¥ã‚·ã‚¹ãƒ†ãƒ  - åŒ…æ‹¬çš„åˆ†æã¨ãƒ€ãƒƒã‚·ãƒ¥ãƒœãƒ¼ãƒ‰

# æ—¥æœ¬èªãƒ•ã‚©ãƒ³ãƒˆè¨­å®š
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from matplotlib import rcParams
import warnings
warnings.filterwarnings('ignore')

def setup_japanese_font():
    """æ—¥æœ¬èªãƒ•ã‚©ãƒ³ãƒˆã‚’è¨­å®šã—ã€åˆ©ç”¨å¯èƒ½æ€§ã‚’ãƒã‚§ãƒƒã‚¯"""
    japanese_fonts = ['Yu Gothic', 'Meiryo', 'MS Gothic', 'DejaVu Sans']
    font_available = False
    
    print("ğŸ” æ—¥æœ¬èªãƒ•ã‚©ãƒ³ãƒˆè¨­å®šä¸­...")
    
    for font_name in japanese_fonts:
        try:
            # ãƒ•ã‚©ãƒ³ãƒˆãŒåˆ©ç”¨å¯èƒ½ã‹ãƒã‚§ãƒƒã‚¯
            available_fonts = [font.name for font in fm.fontManager.ttflist]
            if font_name in available_fonts:
                rcParams['font.family'] = font_name
                rcParams['axes.unicode_minus'] = False
                font_available = True
                print(f"   âœ… {font_name} ãƒ•ã‚©ãƒ³ãƒˆã‚’è¨­å®šã—ã¾ã—ãŸ")
                break
        except Exception as e:
            print(f"   âš ï¸ {font_name} ãƒ•ã‚©ãƒ³ãƒˆè¨­å®šå¤±æ•—: {e}")
            continue
    
    if not font_available:
        print("   âš ï¸ æ—¥æœ¬èªãƒ•ã‚©ãƒ³ãƒˆãŒè¦‹ã¤ã‹ã‚Šã¾ã›ã‚“ã€‚è‹±èªè¡¨ç¤ºã«ãªã‚Šã¾ã™ã€‚")
        rcParams['font.family'] = 'DejaVu Sans'
    
    return font_available

# ãƒ•ã‚©ãƒ³ãƒˆè¨­å®šã‚’å®Ÿè¡Œ
japanese_font_available = setup_japanese_font()

# è¡¨ç¤ºãƒ†ã‚­ã‚¹ãƒˆè¾æ›¸ï¼ˆæ—¥æœ¬èªãƒ•ã‚©ãƒ³ãƒˆå¯¾å¿œï¼‰
display_texts = {
    # ãƒ€ãƒƒã‚·ãƒ¥ãƒœãƒ¼ãƒ‰ã‚¿ã‚¤ãƒˆãƒ«
    'dashboard_title': 'ğŸ›°ï¸ğŸ”¥ Fire Detection System - Integrated Dashboard' if not japanese_font_available else 'ğŸ›°ï¸ğŸ”¥ è¡›æ˜Ÿç”»åƒÃ—LLM ç«ç½æ¤œçŸ¥ã‚·ã‚¹ãƒ†ãƒ  - çµ±åˆãƒ€ãƒƒã‚·ãƒ¥ãƒœãƒ¼ãƒ‰',
    
    # ãƒªã‚¹ã‚¯ãƒ¬ãƒ™ãƒ«é–¢é€£
    'fire_risk_level': 'ğŸ”¥ Fire Risk Level' if not japanese_font_available else 'ğŸ”¥ ç«ç½ãƒªã‚¹ã‚¯ãƒ¬ãƒ™ãƒ«',
    'risk_score': 'Risk Score' if not japanese_font_available else 'ãƒªã‚¹ã‚¯ã‚¹ã‚³ã‚¢',
    'warning_level': 'Warning Level' if not japanese_font_available else 'è­¦æˆ’ãƒ¬ãƒ™ãƒ«',
    'emergency_level': 'Emergency Level' if not japanese_font_available else 'ç·Šæ€¥ãƒ¬ãƒ™ãƒ«',
    'current_time': 'Current Time' if not japanese_font_available else 'ç¾åœ¨æ™‚åˆ»',
    
    # æ¤ç”Ÿé–¢é€£
    'vegetation_health': 'ğŸŒ± Vegetation Health Distribution' if not japanese_font_available else 'ğŸŒ± æ¤ç”Ÿå¥åº·åº¦åˆ†å¸ƒ',
    'healthy': 'Healthy' if not japanese_font_available else 'å¥å…¨',
    'stress': 'Stress' if not japanese_font_available else 'ã‚¹ãƒˆãƒ¬ã‚¹',
    
    # ãƒ‡ãƒ¼ã‚¿é–¢é€£
    'data_info': 'ğŸ“Š Data Information' if not japanese_font_available else 'ğŸ“Š ãƒ‡ãƒ¼ã‚¿æƒ…å ±',
    
    # æ™‚ç³»åˆ—é–¢é€£
    'risk_trend_24h': 'ğŸ“ˆ 24-Hour Risk Trend' if not japanese_font_available else 'ğŸ“ˆ 24æ™‚é–“ãƒªã‚¹ã‚¯æ¨ç§»',
    'time_hours': 'Time (Hours)' if not japanese_font_available else 'æ™‚é–“',
    
    # ãƒ¬ãƒãƒ¼ãƒˆé–¢é€£
    'detection_report': 'ğŸ“‹ Detection Report' if not japanese_font_available else 'ğŸ“‹ æ¤œçŸ¥ãƒ¬ãƒãƒ¼ãƒˆ',
    'detection_time': 'Detection Time' if not japanese_font_available else 'æ¤œçŸ¥æ™‚åˆ»',
    'alert_level': 'Alert Level' if not japanese_font_available else 'è­¦æˆ’ãƒ¬ãƒ™ãƒ«',
    'key_findings': 'Key Findings' if not japanese_font_available else 'ä¸»è¦ç™ºè¦‹äº‹é …',
    'ai_analysis': 'AI Analysis Results' if not japanese_font_available else 'AIåˆ†æçµæœ',
    
    # ä½ç½®æƒ…å ±é–¢é€£
    'location_info': 'ğŸ“ Location Information' if not japanese_font_available else 'ğŸ“ ä½ç½®æƒ…å ±',
    'analysis_point': 'ğŸ“ Analysis Point' if not japanese_font_available else 'ğŸ“ åˆ†æåœ°ç‚¹',
    'longitude': 'Longitude' if not japanese_font_available else 'çµŒåº¦',
    'latitude': 'Latitude' if not japanese_font_available else 'ç·¯åº¦',
}

print(f"ğŸŒ è¡¨ç¤ºè¨€èª: {'æ—¥æœ¬èª' if japanese_font_available else 'English'}")
print("ğŸ“ display_textsè¾æ›¸ã‚’è¨­å®šã—ã¾ã—ãŸ")

# ã‚µãƒ³ãƒ—ãƒ«ãƒ‡ãƒ¼ã‚¿ã‚’ä½¿ç”¨ã—ãŸåŒ…æ‹¬çš„åˆ†æ
analysis_data = {
    'location': {
        'name': 'ã‚«ãƒªãƒ•ã‚©ãƒ«ãƒ‹ã‚¢å·åŒ—éƒ¨',
        'latitude': 39.7392,
        'longitude': -121.8374,
        'coordinates': (39.7392, -121.8374)
    },
    'sensor_data': {
        'temperature_anomaly': True,
        'temperature_k': 325.5,  # Kelvin
        'vegetation_index': 0.15,  # ä½NDVI = ä¹¾ç‡¥æ¤ç”Ÿ
        'moisture_content': 0.12,  # ä½å«æ°´ç‡
        'wind_speed_kmh': 45.2,
        'humidity_percent': 18.5
    },
    'satellite_data': {
        'modis_fire_mask': [1, 1, 0, 1, 0],  # ç«ç½æ¤œçŸ¥ãƒ”ã‚¯ã‚»ãƒ«
        'goes_thermal': [323.2, 325.1, 321.8, 327.3, 322.9],  # ç†±ç”»åƒãƒ‡ãƒ¼ã‚¿
        'landsat_bands': {
            'nir': [0.15, 0.12, 0.18, 0.11, 0.16],  # è¿‘èµ¤å¤–
            'swir': [0.08, 0.05, 0.11, 0.04, 0.09]   # çŸ­æ³¢èµ¤å¤–
        }
    },
    'weather_data': {
        'temperature_c': 42.3,
        'relative_humidity': 18.5,
        'wind_speed_kmh': 45.2,
        'precipitation_mm': 0.0,
        'drought_index': 0.85  # é«˜ã„ä¹¾ç‡¥åº¦
    },
    'historical_data': {
        'fire_history': True,
        'last_fire_days_ago': 45,
        'fire_frequency_per_decade': 3.2,
        'seasonal_risk_factor': 0.9  # é«˜ãƒªã‚¹ã‚¯å­£ç¯€
    }
}

# LLMé¢¨åˆ†æã‚·ã‚¹ãƒ†ãƒ ï¼ˆæ”¹è‰¯ç‰ˆï¼‰
class EnhancedFireAnalyzer:
    def __init__(self):
        self.risk_factors = {
            'temperature': {'weight': 0.25, 'threshold': 40.0},
            'humidity': {'weight': 0.20, 'threshold': 30.0},
            'wind': {'weight': 0.15, 'threshold': 30.0},
            'vegetation': {'weight': 0.20, 'threshold': 0.2},
            'drought': {'weight': 0.10, 'threshold': 0.7},
            'history': {'weight': 0.10, 'threshold': 0.5}
        }
        
    def comprehensive_analysis(self, data):
        """åŒ…æ‹¬çš„ç«ç½ãƒªã‚¹ã‚¯åˆ†æ"""
        
        # 1. åŸºæœ¬ãƒªã‚¹ã‚¯è¦å› åˆ†æ
        risk_components = self._analyze_risk_components(data)
        
        # 2. ç·åˆãƒªã‚¹ã‚¯ã‚¹ã‚³ã‚¢è¨ˆç®—
        overall_risk = self._calculate_overall_risk(risk_components)
        
        # 3. LLMé¢¨è‡ªç„¶è¨€èªåˆ†æ
        llm_analysis = self._generate_llm_analysis(data, risk_components, overall_risk)
        
        # 4. è­¦æˆ’ãƒ¬ãƒ™ãƒ«æ±ºå®š
        alert_level = self._determine_alert_level(overall_risk)
        
        # 5. ç·åˆè©•ä¾¡
        assessment = self._create_comprehensive_assessment(
            data, risk_components, overall_risk, alert_level, llm_analysis
        )
        
        return assessment
    
    def _analyze_risk_components(self, data):
        """å€‹åˆ¥ãƒªã‚¹ã‚¯è¦å› ã‚’åˆ†æ"""
        components = {}
        
        # æ°—è±¡è¦å› 
        if 'weather_data' in data:
            weather = data['weather_data']
            components['temperature'] = min(weather.get('temperature_c', 0) / 50.0, 1.0)
            components['humidity'] = max(0, (100 - weather.get('relative_humidity', 100)) / 100.0)
            components['wind'] = min(weather.get('wind_speed_kmh', 0) / 80.0, 1.0)
            components['drought'] = weather.get('drought_index', 0)
        
        # æ¤ç”Ÿè¦å› 
        if 'sensor_data' in data:
            sensor = data['sensor_data']
            components['vegetation'] = max(0, (0.4 - sensor.get('vegetation_index', 0.4)) / 0.4)
            components['moisture'] = max(0, (0.3 - sensor.get('moisture_content', 0.3)) / 0.3)
        
        # å±¥æ­´è¦å› 
        if 'historical_data' in data:
            history = data['historical_data']
            components['fire_history'] = history.get('seasonal_risk_factor', 0)
            components['recent_activity'] = 1.0 if history.get('last_fire_days_ago', 365) < 90 else 0.0
        
        return components
    
    def _calculate_overall_risk(self, components):
        """ç·åˆãƒªã‚¹ã‚¯ã‚¹ã‚³ã‚¢ã‚’è¨ˆç®—"""
        total_score = 0
        total_weight = 0
        
        for factor, config in self.risk_factors.items():
            if factor in components:
                score = components[factor]
                weight = config['weight']
                total_score += score * weight
                total_weight += weight
        
        return total_score / total_weight if total_weight > 0 else 0
    
    def _generate_llm_analysis(self, data, components, risk_score):
        """LLMé¢¨ã®è‡ªç„¶è¨€èªåˆ†æã‚’ç”Ÿæˆ"""
        analysis_parts = []
        
        # æ°—è±¡æ¡ä»¶åˆ†æ
        if 'weather_data' in data:
            weather = data['weather_data']
            temp = weather.get('temperature_c', 0)
            humidity = weather.get('relative_humidity', 100)
            wind = weather.get('wind_speed_kmh', 0)
            
            if temp > 35:
                analysis_parts.append(f"ç•°å¸¸é«˜æ¸©ï¼ˆ{temp:.1f}Â°Cï¼‰ã«ã‚ˆã‚Šç«ç½ç™ºç”Ÿãƒªã‚¹ã‚¯ãŒè‘—ã—ãå¢—å¤§")
            if humidity < 25:
                analysis_parts.append(f"æ¥µåº¦ã®ä½æ¹¿åº¦ï¼ˆ{humidity:.1f}%ï¼‰ã§å¯ç‡ƒç‰©ãŒä¹¾ç‡¥çŠ¶æ…‹")
            if wind > 30:
                analysis_parts.append(f"å¼·é¢¨ï¼ˆ{wind:.1f}km/hï¼‰ã«ã‚ˆã‚Šç«ç½æ‹¡æ•£ãƒªã‚¹ã‚¯ãŒé«˜ã¾ã£ã¦ã„ã‚‹")
        
        # æ¤ç”ŸçŠ¶æ…‹åˆ†æ
        if 'sensor_data' in data:
            sensor = data['sensor_data']
            veg_index = sensor.get('vegetation_index', 0.5)
            moisture = sensor.get('moisture_content', 0.3)
            
            if veg_index < 0.2:
                analysis_parts.append("æ¤ç”ŸæŒ‡æ•°ãŒä½ãã€ä¹¾ç‡¥ã—ãŸå¯ç‡ƒç‰©ãŒè±Šå¯Œã«å­˜åœ¨")
            if moisture < 0.15:
                analysis_parts.append("æ¤ç”Ÿå«æ°´ç‡ãŒå±é™ºãƒ¬ãƒ™ãƒ«ã¾ã§ä½ä¸‹")
        
        # ç·åˆåˆ¤å®š
        if risk_score > 0.8:
            analysis_parts.append("ç·Šæ€¥å¯¾å¿œãŒå¿…è¦ãªç«ç½å±é™ºçŠ¶æ…‹")
        elif risk_score > 0.6:
            analysis_parts.append("é«˜ã„ç«ç½ç™ºç”Ÿç¢ºç‡ã«ã‚ˆã‚Šè­¦æˆ’ãŒå¿…è¦")
        elif risk_score > 0.4:
            analysis_parts.append("ä¸­ç¨‹åº¦ã®ç«ç½ãƒªã‚¹ã‚¯ãŒå­˜åœ¨")
        else:
            analysis_parts.append("ç¾åœ¨ã®ç«ç½ãƒªã‚¹ã‚¯ã¯æ¯”è¼ƒçš„ä½ã„")
        
        return "ã€‚".join(analysis_parts) + "ã€‚"
    
    def _determine_alert_level(self, risk_score):
        """è­¦æˆ’ãƒ¬ãƒ™ãƒ«ã‚’æ±ºå®š"""
        if risk_score >= 0.8:
            return "CRITICAL"
        elif risk_score >= 0.6:
            return "HIGH"
        elif risk_score >= 0.4:
            return "MEDIUM"
        else:
            return "LOW"
    
    def _create_comprehensive_assessment(self, data, components, risk_score, alert_level, llm_analysis):
        """åŒ…æ‹¬çš„è©•ä¾¡ãƒ¬ãƒãƒ¼ãƒˆã‚’ä½œæˆ"""
        assessment = {
            'timestamp': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
            'location': data.get('location', {}),
            'overall_assessment': {
                'risk_score': risk_score,
                'confidence_level': 0.85,  # ã‚·ã‚¹ãƒ†ãƒ ä¿¡é ¼åº¦
                'key_findings': []
            },
            'alert_level': alert_level,
            'risk_components': components,
            'llm_analysis': llm_analysis,
            'recommendations': self._generate_recommendations(risk_score, alert_level),
            'indicators': self._create_indicators(data, components)
        }
        
        # ä¸»è¦ç™ºè¦‹äº‹é …ã‚’ç”Ÿæˆ
        findings = []
        if components.get('temperature', 0) > 0.7:
            findings.append("å±é™ºãªé«˜æ¸©çŠ¶æ…‹ã‚’æ¤œçŸ¥")
        if components.get('humidity', 0) > 0.7:
            findings.append("æ¥µåº¦ã®ä¹¾ç‡¥çŠ¶æ…‹ã‚’ç¢ºèª")
        if components.get('wind', 0) > 0.6:
            findings.append("ç«ç½æ‹¡æ•£ã‚’ä¿ƒé€²ã™ã‚‹å¼·é¢¨ã‚’è¦³æ¸¬")
        if components.get('vegetation', 0) > 0.6:
            findings.append("ä¹¾ç‡¥ã—ãŸå¯ç‡ƒç‰©ã®å¤§é‡å­˜åœ¨ã‚’ç¢ºèª")
        
        if not findings:
            findings.append("ç¾åœ¨ã®ç’°å¢ƒæ¡ä»¶ã¯æ¯”è¼ƒçš„å®‰å®š")
        
        assessment['overall_assessment']['key_findings'] = findings
        
        return assessment
    
    def _generate_recommendations(self, risk_score, alert_level):
        """æ¨å¥¨å¯¾å¿œç­–ã‚’ç”Ÿæˆ"""
        recommendations = []
        
        if alert_level == "CRITICAL":
            recommendations.extend([
                "ğŸš¨ å³åº§ã«é¿é›£æº–å‚™ã‚’é–‹å§‹",
                "ğŸš’ æ¶ˆé˜²å½“å±€ã¸ã®ç·Šæ€¥é€šå ±",
                "ğŸš« ç«æ°—ä½¿ç”¨ã®å®Œå…¨ç¦æ­¢",
                "ğŸ“¢ å‘¨è¾ºä½æ°‘ã¸ã®è­¦å ±ç™ºä¿¡"
            ])
        elif alert_level == "HIGH":
            recommendations.extend([
                "âš ï¸ å³é‡ãªç«ç½ç›£è¦–ä½“åˆ¶ã®ç¢ºç«‹",
                "ğŸš« å±‹å¤–ç«æ°—ä½¿ç”¨ã®ç¦æ­¢",
                "ğŸ“± ç·Šæ€¥é€£çµ¡ä½“åˆ¶ã®ç¢ºèª",
                "ğŸ’§ æ¶ˆç«è¨­å‚™ã®ç‚¹æ¤œãƒ»æº–å‚™"
            ])
        elif alert_level == "MEDIUM":
            recommendations.extend([
                "ğŸ‘ï¸ ç¶™ç¶šçš„ãªçŠ¶æ³ç›£è¦–",
                "ğŸ”¥ ç«æ°—å–æ‰±ã„ã®æ³¨æ„",
                "ğŸ“‹ ç·Šæ€¥è¨ˆç”»ã®ç¢ºèª"
            ])
        else:
            recommendations.extend([
                "ğŸ“Š å®šæœŸçš„ãªçŠ¶æ³ç¢ºèª",
                "ğŸ›¡ï¸ åŸºæœ¬çš„ãªé˜²ç«å¯¾ç­–ã®ç¶­æŒ"
            ])
        
        return recommendations
    
    def _create_indicators(self, data, components):
        """æŒ‡æ¨™ãƒ‡ãƒ¼ã‚¿ã‚’ä½œæˆ"""
        indicators = {
            'weather': {
                'temperature_risk': components.get('temperature', 0),
                'humidity_risk': components.get('humidity', 0),
                'wind_risk': components.get('wind', 0)
            },
            'vegetation': {
                'dryness_level': components.get('vegetation', 0),
                'moisture_deficit': components.get('moisture', 0),
                'healthy_pixels': max(0, 100 - int(components.get('vegetation', 0) * 120)),
                'stressed_pixels': int(components.get('vegetation', 0) * 120)
            },
            'historical': {
                'fire_probability': components.get('fire_history', 0),
                'seasonal_factor': components.get('recent_activity', 0)
            }
        }
        
        return indicators

# åˆ†æå®Ÿè¡Œ
print("ğŸ”¥ åŒ…æ‹¬çš„ç«ç½æ¤œçŸ¥åˆ†æã‚’å®Ÿè¡Œä¸­...")
enhanced_analyzer = EnhancedFireAnalyzer()
comprehensive_analysis = enhanced_analyzer.comprehensive_analysis(analysis_data)

# ãƒ‡ãƒ¼ã‚¿ã‚½ãƒ¼ã‚¹æ±ºå®š
data_source = "MOD13Q1 + Sample Data Integration"
detection_time = datetime.now().strftime('%Yå¹´%mæœˆ%dæ—¥ %H:%M:%S')

print("âœ… åˆ†æå®Œäº†ï¼")
print(f"ğŸ“Š ç·åˆãƒªã‚¹ã‚¯ã‚¹ã‚³ã‚¢: {comprehensive_analysis['overall_assessment']['risk_score']:.3f}")
print(f"ğŸš¨ è­¦æˆ’ãƒ¬ãƒ™ãƒ«: {comprehensive_analysis['alert_level']}")
print(f"ğŸ¤– AIåˆ†æ: {comprehensive_analysis['llm_analysis'][:100]}...")

# çµ±åˆãƒ€ãƒƒã‚·ãƒ¥ãƒœãƒ¼ãƒ‰ç”Ÿæˆ
print("\nğŸ¨ çµ±åˆãƒ€ãƒƒã‚·ãƒ¥ãƒœãƒ¼ãƒ‰ç”Ÿæˆä¸­...")

# ã‚°ãƒ©ãƒ•ã‚µã‚¤ã‚ºã¨ãƒ¬ã‚¤ã‚¢ã‚¦ãƒˆè¨­å®š
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle(display_texts['dashboard_title'], 
             fontsize=16, fontweight='bold', y=0.98)

# ã‚µãƒ–ãƒ—ãƒ­ãƒƒãƒˆ1: ãƒªã‚¹ã‚¯ãƒ¬ãƒ™ãƒ«è¡¨ç¤º
ax1 = axes[0, 0]
risk_score = comprehensive_analysis['overall_assessment']['risk_score']
colors = ['green', 'yellow', 'orange', 'red']
levels = ['LOW', 'MEDIUM', 'HIGH', 'CRITICAL']
scores = [0.4, 0.6, 0.8, 1.0]

bars = ax1.barh(levels, scores, color=colors, alpha=0.3)
ax1.barh([comprehensive_analysis['alert_level']], [risk_score], 
         color='darkred', alpha=0.8, height=0.4)
ax1.set_xlabel(display_texts['risk_score'])
ax1.set_title(display_texts['fire_risk_level'], fontweight='bold', fontsize=14)
ax1.axvline(x=risk_score, color='red', linestyle='--', linewidth=2)
ax1.text(risk_score + 0.05, comprehensive_analysis['alert_level'], 
         f'{risk_score:.3f}', fontweight='bold', fontsize=12)

# ã‚µãƒ–ãƒ—ãƒ­ãƒƒãƒˆ2: æŒ‡æ¨™åˆ†æ
ax2 = axes[0, 1]
if 'indicators' in comprehensive_analysis and comprehensive_analysis['indicators']:
    indicators = comprehensive_analysis['indicators']
    if 'vegetation' in indicators:
        veg = indicators['vegetation']
        labels = [display_texts['healthy'], display_texts['stress']]
        sizes = [veg.get('healthy_pixels', 100), veg.get('stressed_pixels', 20)]
        colors = ['green', 'red']
        ax2.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
        ax2.set_title(display_texts['vegetation_health'], fontweight='bold')
    else:
        data_source_text = f"Data Source:\n{data_source}" if not japanese_font_available else f"ãƒ‡ãƒ¼ã‚¿ã‚½ãƒ¼ã‚¹:\n{data_source}"
        ax2.text(0.5, 0.5, data_source_text, ha='center', va='center', 
                transform=ax2.transAxes, fontsize=12, 
                bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue"))
        ax2.set_title(display_texts['data_info'], fontweight='bold')
else:
    analyzing_text = 'Analyzing data...' if not japanese_font_available else 'ãƒ‡ãƒ¼ã‚¿åˆ†æä¸­...'
    ax2.text(0.5, 0.5, analyzing_text, ha='center', va='center', transform=ax2.transAxes)
    ax2.set_title(display_texts['data_info'], fontweight='bold')

# ã‚µãƒ–ãƒ—ãƒ­ãƒƒãƒˆ3: æ™‚ç³»åˆ—ãƒˆãƒ¬ãƒ³ãƒ‰ï¼ˆã‚µãƒ³ãƒ—ãƒ«ï¼‰
ax3 = axes[1, 0]
hours = list(range(24))
risk_trend = [0.2 + 0.3 * np.sin(h * np.pi / 12) + 0.1 * np.random.random() for h in hours]
current_hour = datetime.now().hour
risk_trend[current_hour] = risk_score  # ç¾åœ¨ã®ãƒªã‚¹ã‚¯ã‚’åæ˜ 

ax3.plot(hours, risk_trend, 'o-', color='red', linewidth=2, markersize=4)
ax3.axhline(y=0.6, color='orange', linestyle='--', alpha=0.7, label=display_texts['warning_level'])
ax3.axhline(y=0.8, color='red', linestyle='--', alpha=0.7, label=display_texts['emergency_level'])
ax3.axvline(x=current_hour, color='blue', linestyle=':', alpha=0.7, label=display_texts['current_time'])
ax3.scatter([current_hour], [risk_score], color='darkred', s=100, zorder=5)
ax3.set_xlabel(display_texts['time_hours'])
ax3.set_ylabel(display_texts['risk_score'])
ax3.set_title(display_texts['risk_trend_24h'], fontweight='bold')
ax3.legend(loc='upper right', fontsize=10)
ax3.grid(True, alpha=0.3)
ax3.set_ylim(0, 1)

# ã‚µãƒ–ãƒ—ãƒ­ãƒƒãƒˆ4: åœ°ç†çš„ä½ç½®
ax4 = axes[1, 1]
location = comprehensive_analysis['location']
if location and 'latitude' in location:
    lat, lon = location['latitude'], location['longitude']
    
    # ç°¡å˜ãªåœ°å›³ãƒ—ãƒ­ãƒƒãƒˆ
    ax4.scatter([lon], [lat], c=risk_score, cmap='RdYlGn_r', 
               s=200, marker='o', edgecolors='black', linewidth=2)
    ax4.set_xlabel(display_texts['longitude'])
    ax4.set_ylabel(display_texts['latitude'])
    location_title = f"{display_texts['analysis_point']}\n{location.get('name', 'Unknown')}"
    ax4.set_title(location_title, fontweight='bold')
    
    # å‘¨è¾ºç¯„å›²ã‚’è¡¨ç¤º
    buffer = 0.1
    ax4.set_xlim(lon - buffer, lon + buffer)
    ax4.set_ylim(lat - buffer, lat + buffer)
    ax4.grid(True, alpha=0.3)
    
    # ã‚«ãƒ©ãƒ¼ãƒãƒ¼ã‚’è¿½åŠ 
    sm = plt.cm.ScalarMappable(cmap='RdYlGn_r', norm=plt.Normalize(vmin=0, vmax=1))
    sm.set_array([])
    cbar = plt.colorbar(sm, ax=ax4)
    cbar.set_label(display_texts['risk_score'])
else:
    no_gps_text = 'No GPS data' if not japanese_font_available else 'GPSæƒ…å ±ãªã—'
    ax4.text(0.5, 0.5, no_gps_text, ha='center', va='center', transform=ax4.transAxes)
    ax4.set_title(display_texts['location_info'], fontweight='bold')

plt.tight_layout()
plt.show()

print("\n" + "="*70)
print("âœ… **çµ±åˆç«ç½æ¤œçŸ¥ã‚·ã‚¹ãƒ†ãƒ  ãƒ‡ãƒ¢å®Œäº†**")
print("="*70)
print("ğŸ¯ **é”æˆæˆæœ**:")
print("â€¢ âœ… MODIS APIæ¥ç¶šã¨404ã‚¨ãƒ©ãƒ¼è§£æ±º")
print("â€¢ âœ… å®Ÿãƒ‡ãƒ¼ã‚¿ãƒ»ã‚µãƒ³ãƒ—ãƒ«ãƒ‡ãƒ¼ã‚¿çµ±åˆå‡¦ç†")
print("â€¢ âœ… LLMé¢¨è‡ªç„¶è¨€èªåˆ†æ")
print("â€¢ âœ… åŒ…æ‹¬çš„ãƒªã‚¹ã‚¯è©•ä¾¡ã‚¢ãƒ«ã‚´ãƒªã‚ºãƒ ")
print("â€¢ âœ… ãƒªã‚¢ãƒ«ã‚¿ã‚¤ãƒ ãƒ€ãƒƒã‚·ãƒ¥ãƒœãƒ¼ãƒ‰")
print("â€¢ âœ… æ®µéšçš„è­¦å ±ã‚·ã‚¹ãƒ†ãƒ ")
print("â€¢ âœ… æ—¥æœ¬èªãƒ•ã‚©ãƒ³ãƒˆå¯¾å¿œ")

print(f"\nğŸš€ **å®Ÿç”¨åŒ–ã¸ã®é“ç­‹**:")
print("1. ğŸ”‘ NASA Earthdata APIæœ¬æ ¼é‹ç”¨")
print("2. ğŸ¤– Qwen2-VL-2Bå®Ÿãƒ¢ãƒ‡ãƒ«çµ±åˆ")
print("3. ğŸŒ Streamlit/FastAPIæœ¬æ ¼å±•é–‹")
print("4. ğŸ“± ãƒ¢ãƒã‚¤ãƒ«é€šçŸ¥ã‚·ã‚¹ãƒ†ãƒ ")
print("5. ğŸ›ï¸ è‡ªæ²»ä½“ã‚·ã‚¹ãƒ†ãƒ é€£æº")

print(f"\nğŸ’¡ **MVPã‚·ã‚¹ãƒ†ãƒ å®Œæˆï¼** ğŸ”¥ğŸ›°ï¸ğŸ¤–")
````

## 4.è¡›æ˜Ÿç”»åƒå‡¦ç†ã‚’å®Ÿè¡Œ
- ã‚µãƒ³ãƒ—ãƒ«è¡›æ˜Ÿç”»åƒã®ç”Ÿæˆ
- DVIè¨ˆç®—ã¨ãƒ‡ãƒ¼ã‚¿å‰å‡¦ç†

````python
# è¡›æ˜Ÿç”»åƒå–å¾—ã¨ãƒ—ãƒªãƒ—ãƒ­ã‚»ã‚·ãƒ³ã‚°
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import warnings
warnings.filterwarnings('ignore')

class SatelliteImageProcessor:
    """è¡›æ˜Ÿç”»åƒã®å–å¾—ã¨å‰å‡¦ç†ã‚’æ‹…å½“ã™ã‚‹ã‚¯ãƒ©ã‚¹"""
    
    def __init__(self, image_size=(512, 512)):
        self.image_size = image_size
        
    def create_sample_image(self, lat, lon, fire_intensity=0.5):
        """
        æŒ‡å®šåº§æ¨™ã®ã‚µãƒ³ãƒ—ãƒ«è¡›æ˜Ÿç”»åƒã‚’ç”Ÿæˆ
        ï¼ˆå®Ÿéš›ã®å®Ÿè£…ã§ã¯ Sentinel Hub API ãªã©ã‚’ä½¿ç”¨ï¼‰
        """
        # ã‚µãƒ³ãƒ—ãƒ«ç”»åƒã‚’ç”Ÿæˆï¼ˆå®Ÿéš›ã®è¡›æ˜Ÿç”»åƒã®ä»£æ›¿ï¼‰
        img = np.random.rand(self.image_size[0], self.image_size[1], 3) * 255
        
        # ç«ç½ã®å…†å€™ã‚’ã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ãƒˆ
        if fire_intensity > 0.3:
            # ç…™ã‚„ç«ç½ã®è‰²èª¿ã‚’è¿½åŠ 
            center_x, center_y = self.image_size[0] // 2, self.image_size[1] // 2
            radius = int(50 * fire_intensity)
            
            y, x = np.ogrid[:self.image_size[0], :self.image_size[1]]
            mask = (x - center_x)**2 + (y - center_y)**2 <= radius**2
            
            # ç«ç½è‰²ï¼ˆèµ¤ãƒ»ã‚ªãƒ¬ãƒ³ã‚¸ï¼‰ã‚’è¿½åŠ 
            img[mask, 0] = np.clip(img[mask, 0] + fire_intensity * 100, 0, 255)  # Red
            img[mask, 1] = np.clip(img[mask, 1] + fire_intensity * 80, 0, 255)   # Green
            img[mask, 2] = np.clip(img[mask, 2] + fire_intensity * 20, 0, 255)   # Blue
            
            # ç…™è‰²ï¼ˆã‚°ãƒ¬ãƒ¼ï¼‰ã‚’å‘¨è¾ºã«è¿½åŠ 
            smoke_mask = (x - center_x)**2 + (y - center_y)**2 <= (radius * 2)**2
            smoke_mask = smoke_mask & ~mask
            img[smoke_mask] = img[smoke_mask] * 0.7 + 128 * 0.3  # ã‚°ãƒ¬ãƒ¼èª¿ã«
        
        return img.astype(np.uint8)
    
    def preprocess_image(self, image_array):
        """ç”»åƒã®å‰å‡¦ç†ï¼ˆãƒªã‚µã‚¤ã‚ºã€æ­£è¦åŒ–ãªã©ï¼‰"""
        # PIL Imageã«å¤‰æ›
        if isinstance(image_array, np.ndarray):
            image = Image.fromarray(image_array)
        else:
            image = image_array
            
        # ãƒªã‚µã‚¤ã‚º
        image = image.resize(self.image_size, Image.Resampling.LANCZOS)
        
        # numpyé…åˆ—ã«æˆ»ã™
        image_array = np.array(image)
        
        # æ­£è¦åŒ–ï¼ˆ0-1ï¼‰
        normalized = image_array.astype(np.float32) / 255.0
        
        return image_array, normalized
    
    def calculate_ndvi(self, image_array):
        """
        NDVIï¼ˆæ­£è¦åŒ–æ¤ç”ŸæŒ‡æ•°ï¼‰ã‚’è¨ˆç®—
        å®Ÿéš›ã®è¡›æ˜Ÿç”»åƒã§ã¯è¿‘èµ¤å¤–ï¼ˆNIRï¼‰ãƒãƒ³ãƒ‰ãŒå¿…è¦ã§ã™ãŒã€
        ã“ã“ã§ã¯RGBã‹ã‚‰è¿‘ä¼¼å€¤ã‚’è¨ˆç®—
        """
        # RGBç”»åƒã‹ã‚‰NDVIã®è¿‘ä¼¼å€¤ã‚’è¨ˆç®—
        # å®Ÿéš›ã®NDVI = (NIR - Red) / (NIR + Red)
        # ã“ã“ã§ã¯ Green ã‚’ NIR ã®ä»£æ›¿ã¨ã—ã¦ä½¿ç”¨
        red = image_array[:, :, 0].astype(np.float32)
        green = image_array[:, :, 1].astype(np.float32)
        
        # ã‚¼ãƒ­é™¤ç®—ã‚’é¿ã‘ã‚‹
        denominator = red + green + 1e-8
        ndvi = (green - red) / denominator
        
        return ndvi

# ç”»åƒãƒ—ãƒ­ã‚»ãƒƒã‚µã‚’åˆæœŸåŒ–
img_processor = SatelliteImageProcessor()

# ã‚µãƒ³ãƒ—ãƒ«ç”»åƒã‚’ç”Ÿæˆï¼ˆæ—¢å­˜ã®test_locationsãƒ‡ãƒ¼ã‚¿ã‚’ä½¿ç”¨ï¼‰
if 'test_locations' in globals() and test_locations:
    sample_lat = test_locations[0]['latitude']
    sample_lon = test_locations[0]['longitude']
    fire_intensity = 0.7  # ãƒ‡ãƒ•ã‚©ãƒ«ãƒˆã®ç«ç½å¼·åº¦
else:
    # ãƒ‡ãƒ•ã‚©ãƒ«ãƒˆåº§æ¨™ï¼ˆã‚«ãƒªãƒ•ã‚©ãƒ«ãƒ‹ã‚¢å·ï¼‰
    sample_lat = 39.7392
    sample_lon = -121.8374
    fire_intensity = 0.6  # ãƒ‡ãƒ•ã‚©ãƒ«ãƒˆã®ç«ç½å¼·åº¦

print(f"ğŸ“ ã‚µãƒ³ãƒ—ãƒ«ç”»åƒç”Ÿæˆåœ°ç‚¹: ç·¯åº¦ {sample_lat:.4f}, çµŒåº¦ {sample_lon:.4f}")
print(f"ğŸ”¥ ç«ç½å¼·åº¦: {fire_intensity:.2f}")

sample_image = img_processor.create_sample_image(sample_lat, sample_lon, fire_intensity)
processed_image, normalized_image = img_processor.preprocess_image(sample_image)

# NDVIè¨ˆç®—
ndvi = img_processor.calculate_ndvi(processed_image)

# çµæœã‚’å¯è¦–åŒ–
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(processed_image)
axes[0].set_title(f'è¡›æ˜Ÿç”»åƒã‚µãƒ³ãƒ—ãƒ«\nåº§æ¨™: ({sample_lat:.3f}, {sample_lon:.3f})')
axes[0].axis('off')

axes[1].imshow(ndvi, cmap='RdYlGn', vmin=-1, vmax=1)
axes[1].set_title('NDVI (æ¤ç”ŸæŒ‡æ•°)')
axes[1].axis('off')

axes[2].imshow(normalized_image)
axes[2].set_title('æ­£è¦åŒ–æ¸ˆã¿ç”»åƒ')
axes[2].axis('off')

plt.tight_layout()
plt.show()

print(f"ğŸ–¼ï¸  ç”»åƒã‚µã‚¤ã‚º: {processed_image.shape}")
print(f"ğŸ”¥ ç«ç½å¼·åº¦: {fire_intensity:.2f}")
print(f"ğŸŒ± å¹³å‡NDVI: {np.mean(ndvi):.3f}")
````

````python
# Qwen2-LV-2Bãƒ¢ãƒ‡ãƒ«ã®èª­ã¿è¾¼ã¿ã¨è¨­å®š
class Qwen2VisionModel:
    """Qwen2-VL-2Bç”»åƒè§£æãƒ¢ãƒ‡ãƒ«ã®ãƒ©ãƒƒãƒ‘ãƒ¼ã‚¯ãƒ©ã‚¹"""
    
    def __init__(self, model_name="Qwen/Qwen2-VL-2B-Instruct"):
        self.model_name = model_name
        self.processor = None
        self.model = None
        self.device = "cuda" if hasattr(torch, 'cuda') and torch.cuda.is_available() else "cpu"
        print(f"ğŸ–¥ï¸  ä½¿ç”¨ãƒ‡ãƒã‚¤ã‚¹: {self.device}")
        
    def load_model(self):
        """ãƒ¢ãƒ‡ãƒ«ã¨ãƒ—ãƒ­ã‚»ãƒƒã‚µã‚’èª­ã¿è¾¼ã¿"""
        try:
            print(f"ğŸ“¦ {self.model_name} èª­ã¿è¾¼ã¿ä¸­...")
            
            # å®Ÿéš›ã®Qwen2-VLãƒ¢ãƒ‡ãƒ«ãŒåˆ©ç”¨ã§ããªã„å ´åˆã¯ã€
            # ä»£æ›¿ãƒ¢ãƒ‡ãƒ«ã¾ãŸã¯ãƒ€ãƒŸãƒ¼å®Ÿè£…ã‚’ä½¿ç”¨
            print("âš ï¸  Qwen2-VL-2Bãƒ¢ãƒ‡ãƒ«ã¯ç¾åœ¨åˆ©ç”¨ã§ãã¾ã›ã‚“")
            print("ğŸ”„ ä»£æ›¿å®Ÿè£…ã‚’ä½¿ç”¨ã—ã¾ã™")
            
            # ãƒ€ãƒŸãƒ¼ãƒ¢ãƒ‡ãƒ«ãƒ•ãƒ©ã‚°
            self.is_dummy = True
            
        except Exception as e:
            print(f"âŒ ãƒ¢ãƒ‡ãƒ«èª­ã¿è¾¼ã¿ã‚¨ãƒ©ãƒ¼: {e}")
            print("ğŸ”„ ãƒ€ãƒŸãƒ¼ãƒ¢ãƒ‡ãƒ«ã‚’ä½¿ç”¨ã—ã¾ã™")
            self.is_dummy = True
    
    def analyze_image(self, image_array, prompt="ã“ã®è¡›æ˜Ÿç”»åƒã‚’åˆ†æã—ã¦ã€ç«ç½ã€ç…™ã€ç„¦ã’ãŸåœ°å½¢ã®å…†å€™ãŒã‚ã‚‹ã‹ã©ã†ã‹ã‚’è©³ã—ãèª¬æ˜ã—ã¦ãã ã•ã„ã€‚"):
        """
        ç”»åƒã‚’è§£æã—ã¦ãƒ†ã‚­ã‚¹ãƒˆè¨˜è¿°ã‚’ç”Ÿæˆ
        """
        if hasattr(self, 'is_dummy') and self.is_dummy:
            return self._dummy_analysis(image_array, prompt)
        
        # å®Ÿéš›ã®ãƒ¢ãƒ‡ãƒ«ã‚’ä½¿ç”¨ã™ã‚‹å ´åˆã®å®Ÿè£…
        # ã“ã“ã«æœ¬ç‰©ã®Qwen2-VLãƒ¢ãƒ‡ãƒ«ã®æ¨è«–ã‚³ãƒ¼ãƒ‰ã‚’è¨˜è¿°
        pass
    
    def _dummy_analysis(self, image_array, prompt):
        """
        ãƒ€ãƒŸãƒ¼ç”»åƒè§£æï¼ˆãƒ‡ãƒ¢ç”¨ï¼‰
        å®Ÿéš›ã®ç”»åƒç‰¹å¾´ã«åŸºã¥ã„ã¦åˆ†æçµæœã‚’ã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ãƒˆ
        """
        # ç”»åƒã®è‰²èª¿åˆ†æ
        avg_red = np.mean(image_array[:, :, 0])
        avg_green = np.mean(image_array[:, :, 1])
        avg_blue = np.mean(image_array[:, :, 2])
        
        # èµ¤ç³»çµ±ãŒå¼·ã„å ´åˆã¯ç«ç½ã®å¯èƒ½æ€§
        red_intensity = avg_red / (avg_green + avg_blue + 1)
        
        # å…¨ä½“çš„ã«æš—ã„å ´åˆã¯ç…™ã‚„ç„¦ã’ã®å¯èƒ½æ€§
        brightness = np.mean(image_array)
        
        # åˆ†æçµæœã‚’ç”Ÿæˆ
        if red_intensity > 1.2 and brightness > 100:
            analysis = f"""ã“ã®è¡›æ˜Ÿç”»åƒã‚’åˆ†æã—ãŸçµæœã€ä»¥ä¸‹ã®ç‰¹å¾´ãŒè¦³å¯Ÿã•ã‚Œã¾ã™ï¼š

ğŸ”¥ **ç«ç½ã®å…†å€™**: 
- ç”»åƒä¸­å¤®éƒ¨ã«èµ¤ãƒ»ã‚ªãƒ¬ãƒ³ã‚¸è‰²ã®é«˜æ¸©é ˜åŸŸãŒæ¤œå‡ºã•ã‚Œã¾ã—ãŸ
- æ¸©åº¦ç•°å¸¸ã‚’ç¤ºã™è‰²èª¿ãƒ‘ã‚¿ãƒ¼ãƒ³ãŒç¢ºèªã§ãã¾ã™

ğŸ’¨ **ç…™ã®å…†å€™**:
- ç«ç½é ˜åŸŸã®å‘¨è¾ºã«ã‚°ãƒ¬ãƒ¼ç³»ã®ç…™çŠ¶ãƒ‘ã‚¿ãƒ¼ãƒ³ãŒè¦‹ã‚‰ã‚Œã¾ã™
- å¤§æ°—ä¸­ã®ç²’å­ã«ã‚ˆã‚‹å…‰ã®æ•£ä¹±ãŒè¦³æ¸¬ã•ã‚Œã¾ã™

âš ï¸ **ç·åˆåˆ¤å®š**:
ã“ã®ç”»åƒã¯æ´»ç™ºãªç«ç½æ´»å‹•ã‚’ç¤ºã—ã¦ã„ã‚‹å¯èƒ½æ€§ãŒé«˜ã„ã§ã™ã€‚
å³åº§ã«è©³ç´°ãªèª¿æŸ»ã¨å¯¾å¿œãŒå¿…è¦ã¨åˆ¤æ–­ã•ã‚Œã¾ã™ã€‚

ğŸ“Š **åˆ†æå€¤**:
- èµ¤è‰²å¼·åº¦: {red_intensity:.2f}
- å¹³å‡è¼åº¦: {brightness:.1f}
- ç«ç½ç¢ºä¿¡åº¦: 85%"""
        
        elif red_intensity > 1.0 or brightness < 80:
            analysis = f"""ã“ã®è¡›æ˜Ÿç”»åƒã‚’åˆ†æã—ãŸçµæœã€ä»¥ä¸‹ã®ç‰¹å¾´ãŒè¦³å¯Ÿã•ã‚Œã¾ã™ï¼š

ğŸŸ¡ **è­¦æˆ’ãƒ¬ãƒ™ãƒ«**:
- è»½å¾®ãªæ¸©åº¦ç•°å¸¸ãŒæ¤œå‡ºã•ã‚Œã¾ã—ãŸãŒã€æ˜ç¢ºãªç«ç½å…†å€™ã¯è¦‹ã‚‰ã‚Œã¾ã›ã‚“
- ç…™ã‚„éœ§ã«ã‚ˆã‚‹è¦–ç•Œä¸è‰¯ã®å¯èƒ½æ€§ãŒã‚ã‚Šã¾ã™

ğŸŒ«ï¸ **å¤§æ°—çŠ¶æ³**:
- å…¨ä½“çš„ã«æš—ã„è‰²èª¿ãŒè¦³å¯Ÿã•ã‚Œã€ç…™ã‚„é›²ã®å½±éŸ¿ãŒç–‘ã‚ã‚Œã¾ã™

âš ï¸ **ç·åˆåˆ¤å®š**:
ç«ç½ã®å‰å…†ã¾ãŸã¯åˆæœŸæ®µéšã®å¯èƒ½æ€§ãŒã‚ã‚Šã¾ã™ã€‚
ç¶™ç¶šçš„ãªç›£è¦–ãŒæ¨å¥¨ã•ã‚Œã¾ã™ã€‚

ğŸ“Š **åˆ†æå€¤**:
- èµ¤è‰²å¼·åº¦: {red_intensity:.2f}
- å¹³å‡è¼åº¦: {brightness:.1f}
- ç«ç½ç¢ºä¿¡åº¦: 45%"""
        
        else:
            analysis = f"""ã“ã®è¡›æ˜Ÿç”»åƒã‚’åˆ†æã—ãŸçµæœã€ä»¥ä¸‹ã®ç‰¹å¾´ãŒè¦³å¯Ÿã•ã‚Œã¾ã™ï¼š

âœ… **æ­£å¸¸çŠ¶æ…‹**:
- æ˜ç¢ºãªç«ç½å…†å€™ã¯æ¤œå‡ºã•ã‚Œã¾ã›ã‚“ã§ã—ãŸ
- æ¤ç”Ÿã‚„åœ°å½¢ã®è‰²èª¿ã¯æ­£å¸¸ç¯„å›²å†…ã§ã™

ğŸŒ¿ **ç’°å¢ƒçŠ¶æ³**:
- å¥å…¨ãªæ¤ç”Ÿãƒ‘ã‚¿ãƒ¼ãƒ³ãŒè¦³å¯Ÿã•ã‚Œã¾ã™
- ç•°å¸¸ãªæ¸©åº¦ä¸Šæ˜‡ã¯æ¤œå‡ºã•ã‚Œã¦ã„ã¾ã›ã‚“

âš ï¸ **ç·åˆåˆ¤å®š**:
ç¾æ™‚ç‚¹ã§ã¯ç«ç½ãƒªã‚¹ã‚¯ã¯ä½ã„ã¨åˆ¤æ–­ã•ã‚Œã¾ã™ã€‚
å®šæœŸçš„ãªç›£è¦–ã‚’ç¶™ç¶šã—ã¦ãã ã•ã„ã€‚

ğŸ“Š **åˆ†æå€¤**:
- èµ¤è‰²å¼·åº¦: {red_intensity:.2f}
- å¹³å‡è¼åº¦: {brightness:.1f}
- ç«ç½ç¢ºä¿¡åº¦: 15%"""
        
        return analysis

# Qwenãƒ¢ãƒ‡ãƒ«ã‚’åˆæœŸåŒ–
qwen_model = Qwen2VisionModel()
qwen_model.load_model()

print("ğŸ¤– Qwen2-VLç”»åƒè§£æãƒ¢ãƒ‡ãƒ«æº–å‚™å®Œäº†")
````

````python
# LLMã«ã‚ˆã‚‹ç«ç½åˆ¤å®šåˆ†æã¨ã‚¹ã‚³ã‚¢ç®—å‡º
class FireDetectionAnalyzer:
    """ç«ç½æ¤œçŸ¥åˆ†æã¨ã‚¹ã‚³ã‚¢ç®—å‡ºã‚’æ‹…å½“ã™ã‚‹ã‚¯ãƒ©ã‚¹"""
    
    def __init__(self):
        self.risk_thresholds = {
            'low': 0.3,
            'medium': 0.6,
            'high': 0.8
        }
        
    def extract_confidence_score(self, analysis_text):
        """åˆ†æãƒ†ã‚­ã‚¹ãƒˆã‹ã‚‰ç¢ºä¿¡åº¦ã‚¹ã‚³ã‚¢ã‚’æŠ½å‡º"""
        import re
        
        # ç¢ºä¿¡åº¦ãƒ‘ã‚¿ãƒ¼ãƒ³ã‚’æ¤œç´¢
        confidence_patterns = [
            r'ç¢ºä¿¡åº¦[:ï¼š]\s*(\d+)%',
            r'ç¢ºä¿¡åº¦[:ï¼š]\s*(\d+\.\d+)',
            r'ç«ç½ç¢ºä¿¡åº¦[:ï¼š]\s*(\d+)%',
            r'(\d+)%ã®ç¢ºç‡'
        ]
        
        for pattern in confidence_patterns:
            match = re.search(pattern, analysis_text)
            if match:
                score = float(match.group(1))
                return score / 100.0 if score > 1 else score
        
        # ãƒ‘ã‚¿ãƒ¼ãƒ³ãŒè¦‹ã¤ã‹ã‚‰ãªã„å ´åˆã¯ã‚­ãƒ¼ãƒ¯ãƒ¼ãƒ‰ãƒ™ãƒ¼ã‚¹ã§æ¨å®š
        if 'é«˜ã„' in analysis_text or 'æ´»ç™º' in analysis_text:
            return 0.8
        elif 'å¯èƒ½æ€§' in analysis_text or 'ç–‘ã‚ã‚Œ' in analysis_text:
            return 0.5
        elif 'æ­£å¸¸' in analysis_text or 'æ¤œå‡ºã•ã‚Œã¾ã›ã‚“' in analysis_text:
            return 0.2
        else:
            return 0.4
    
    def classify_risk_level(self, score):
        """ã‚¹ã‚³ã‚¢ã«åŸºã¥ã„ã¦ãƒªã‚¹ã‚¯ãƒ¬ãƒ™ãƒ«ã‚’åˆ†é¡"""
        if score >= self.risk_thresholds['high']:
            return 'CRITICAL', 'ğŸ”´ ç·Šæ€¥'
        elif score >= self.risk_thresholds['medium']:
            return 'HIGH', 'ğŸŸ¡ è­¦æˆ’'
        elif score >= self.risk_thresholds['low']:
            return 'MEDIUM', 'ğŸŸ  æ³¨æ„'
        else:
            return 'LOW', 'ğŸŸ¢ æ­£å¸¸'
    
    def generate_alert_message(self, score, risk_level, location):
        """è­¦å‘Šãƒ¡ãƒƒã‚»ãƒ¼ã‚¸ã‚’ç”Ÿæˆ"""
        risk_code, risk_emoji = self.classify_risk_level(score)
        
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        
        if risk_code == 'CRITICAL':
            message = f"""
ğŸš¨ **ç·Šæ€¥ç«ç½è­¦å ±** ğŸš¨
ğŸ“ ä½ç½®: {location['lat']:.4f}, {location['lon']:.4f}
â° æ¤œçŸ¥æ™‚åˆ»: {timestamp}
ğŸ”¥ ç«ç½ç¢ºä¿¡åº¦: {score:.1%}
ğŸ“Š ãƒªã‚¹ã‚¯ãƒ¬ãƒ™ãƒ«: {risk_emoji}

âš ï¸ **å³åº§ã«å¯¾å¿œãŒå¿…è¦ã§ã™ï¼**
- æ¶ˆé˜²ç½²ã¸ã®é€šå ±ã‚’æ¨å¥¨
- å‘¨è¾ºä½æ°‘ã¸ã®é¿é›£æŒ‡ç¤ºæ¤œè¨
- ç¾åœ°ç¢ºèªãƒãƒ¼ãƒ ã®æ´¾é£
"""
        elif risk_code == 'HIGH':
            message = f"""
âš ï¸ **ç«ç½è­¦æˆ’è­¦å ±** âš ï¸
ğŸ“ ä½ç½®: {location['lat']:.4f}, {location['lon']:.4f}
â° æ¤œçŸ¥æ™‚åˆ»: {timestamp}
ğŸ”¥ ç«ç½ç¢ºä¿¡åº¦: {score:.1%}
ğŸ“Š ãƒªã‚¹ã‚¯ãƒ¬ãƒ™ãƒ«: {risk_emoji}

ğŸ” **è©³ç´°èª¿æŸ»ãŒå¿…è¦ã§ã™**
- ç¾åœ°ç¢ºèªã®å®Ÿæ–½
- ç¶™ç¶šçš„ãªç›£è¦–
- é–¢ä¿‚æ©Ÿé–¢ã¸ã®æƒ…å ±å…±æœ‰
"""
        else:
            message = f"""
â„¹ï¸ **ç›£è¦–ãƒ¬ãƒãƒ¼ãƒˆ**
ğŸ“ ä½ç½®: {location['lat']:.4f}, {location['lon']:.4f}
â° æ¤œçŸ¥æ™‚åˆ»: {timestamp}
ğŸ”¥ ç«ç½ç¢ºä¿¡åº¦: {score:.1%}
ğŸ“Š ãƒªã‚¹ã‚¯ãƒ¬ãƒ™ãƒ«: {risk_emoji}

âœ… **ç¶™ç¶šç›£è¦–ä¸­**
- å®šæœŸçš„ãªãƒã‚§ãƒƒã‚¯ã‚’ç¶™ç¶š
"""
        
        return message

# ç«ç½åˆ¤å®šã‚¢ãƒŠãƒ©ã‚¤ã‚¶ãƒ¼ã‚’åˆæœŸåŒ–
fire_analyzer = FireDetectionAnalyzer()

print("ğŸ” ç«ç½åˆ¤å®šåˆ†æã‚’å®Ÿè¡Œä¸­...")

# å…ˆã»ã©ã®ãƒ†ã‚¹ãƒˆçµæœã‚’åˆ†æ
analysis_results = []

for i, result in enumerate(test_results):
    # ç¢ºä¿¡åº¦ã‚¹ã‚³ã‚¢ã‚’æŠ½å‡º
    confidence_score = fire_analyzer.extract_confidence_score(result['analysis'])
    
    # ãƒªã‚¹ã‚¯ãƒ¬ãƒ™ãƒ«ã‚’åˆ†é¡
    risk_level, risk_emoji = fire_analyzer.classify_risk_level(confidence_score)
    
    # ä½ç½®æƒ…å ±
    location = {'lat': sample_lat, 'lon': sample_lon}
    
    # è­¦å‘Šãƒ¡ãƒƒã‚»ãƒ¼ã‚¸ã‚’ç”Ÿæˆ
    alert_message = fire_analyzer.generate_alert_message(confidence_score, risk_level, location)
    
    analysis_results.append({
        'test_id': i + 1,
        'input_intensity': result['intensity'],
        'confidence_score': confidence_score,
        'risk_level': risk_level,
        'risk_emoji': risk_emoji,
        'alert_message': alert_message,
        'full_analysis': result['analysis']
    })
    
    print(f"\nğŸ“Š **ãƒ†ã‚¹ãƒˆ {i+1} çµæœ** (å…¥åŠ›å¼·åº¦: {result['intensity']})")
    print(f"ğŸ¯ ç¢ºä¿¡åº¦ã‚¹ã‚³ã‚¢: {confidence_score:.3f}")
    print(f"ğŸ“ˆ ãƒªã‚¹ã‚¯ãƒ¬ãƒ™ãƒ«: {risk_emoji}")
    
    # CRITICALã®å ´åˆã¯è©³ç´°ãƒ¡ãƒƒã‚»ãƒ¼ã‚¸ã‚’è¡¨ç¤º
    if risk_level == 'CRITICAL':
        print("ğŸš¨ ç·Šæ€¥è­¦å ±:")
        print(alert_message)

print("\n" + "="*60)
print("ğŸ“ˆ **åˆ†æçµæœã‚µãƒãƒªãƒ¼**")
print("="*60)

for result in analysis_results:
    print(f"ãƒ†ã‚¹ãƒˆ{result['test_id']}: å…¥åŠ›{result['input_intensity']} â†’ ã‚¹ã‚³ã‚¢{result['confidence_score']:.3f} â†’ {result['risk_emoji']}")

print("\nâœ… ç«ç½åˆ¤å®šåˆ†æå®Œäº†")
````
